```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/superstore.csv')

df.head()
df.info()
df.isnull().sum()
df.duplicated().sum()

# Total Sales
print(df["Sales"].sum())

# Total Profit
print(df["Profit"].sum())

# Average Sales
average_sales = df["Sales"].mean()
print("Average Sales:", average_sales)

# Sales by Category
df.groupby("Category")["Sales"].sum()

# Profit by Category
profit_by_category = df.groupby("Category")["Profit"].sum()
print(profit_by_category)

# Top Cities
df.groupby("City")["Sales"].sum().sort_values(ascending=False).head(10)

# Top Products
top_products = df.groupby("Sub-Category")["Sales"].sum().sort_values(ascending=False).head(10)
print(top_products)

#Region Analysis
region_sales = df.groupby("Region")["Sales"].sum()
print(region_sales)

#Discount Impact
discount_profit = df.groupby("Discount")["Profit"].mean()
print(discount_profit)
```
